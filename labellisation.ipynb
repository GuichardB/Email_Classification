## Import and prepare Data

```python
import numpy as np
import pandas as pd
import ast
```

```python
data = pd.read_csv("datacleaned/df_10000.CSV")
data = data.drop(['Unnamed: 0'], axis=1) #remove columns creted bc of the to_csv() in the cleaning
data = data.iloc[0:10000]
print(data.shape)
```

```python
i = 0
while i < len(data):
    data['objet'][i] = ast.literal_eval(data['objet'][i])
    data['corps'][i] = ast.literal_eval(data['corps'][i])
    data['adresse'][i] = ast.literal_eval(data['adresse'][i])
    i+=1
```

```python
data['cat'] = ''
```

```python
i = 0
data['adressebis'] = ''
while i < len(data):
    if len(data['adresse'][i])==2:
        data['adressebis'][i] = data['adresse'][i][0]+'@'+data['adresse'][i][1]
    else:
        data['adressebis'][i] = data['adresse'][i][0]+'@'+data['adresse'][i][1]+'.'+data['adresse'][i][2]
    i+=1
```

## Creation of functions

```python
def print_cat(df, str):
    mask = (df['cat'] == str)
    display(df[mask])
```

```python
def print_adresses_with_no_cat(df):
    mask = (df['cat'] == '')
    adresses = df[mask]['adressebis'].unique()
    print(len(adresses))
    for i in adresses:
            print(i)
```

```python
# Labelling using email adress
def cat_with_list(df, list, str):
    i=0
    while i < len(df):
        if df['adressebis'][i] in list:
            df['cat'][i] = str
        i+=1

    return df
```

```python
# Labelling with words in main content
def cat_with_words(df, words, str):
    words_bool = []
    x=0
    while x < len(words):
        words_bool.append(False)
        x+=1
    i=0
    while i < len(df):
        x=0
        while x < len(words):
            words_bool[x] = False
            x+=1
        for word in df['corps'][i]:
            x = 0
            while x < len(words):
                if word == words[x]:
                    words_bool[x] = True
                x+=1
        x=0
        c=0
        while x < len(words):
            if words_bool[x] == True:
                c+=1
            x+=1
        if c == len(words):
            df['cat'][i] = str
        i+=1
    
    return df
```

```python
# Labelling with email object
def cat_with_objet(df, str_to_find, str_cat):
    i = 0
    while i < len(df):
        present = False
        for objet in df['objet'][i]:
            if str_to_find in objet:
                present=True
        if present == True:
            df['cat'][i] = str_cat
        i+=1
    return df
```
